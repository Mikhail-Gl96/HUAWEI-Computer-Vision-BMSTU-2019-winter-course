```python
# replace dataset to script directory

# # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
# #                               PATHS SCRIPT
# # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
# path_to_script = abs_path_start + '/060/UGATIT-pytorch-master' + '/main.py'
# path_to_datasets = abs_path_start + '/060/UGATIT-pytorch-master/dataset/'

# path_trainA = path_to_datasets + '/trainA'
# path_trainB = path_to_datasets + '/trainB'

# path_testA = path_to_datasets + '/testA'
# path_testB = path_to_datasets + '/testB'


# your_dataset_name = 'SIMPlified' # input
if your_dataset_name not in os.listdir(path_to_datasets):
    os.mkdir(your_dataset_name)
os.chdir(path_to_datasets)
if f'{your_dataset_name}' not in os.listdir(path_to_datasets):
    os.mkdir(your_dataset_name)
if '__temp' not in os.listdir(path_to_datasets):  # not resized pics 
    os.mkdir('__temp')
if 'temp' not in os.listdir(path_to_datasets):    # resized pics 
    os.mkdir('temp')
    
path_to_datasets_scriptDir = path_to_datasets + f'{your_dataset_name}'

os.chdir(path_to_datasets_scriptDir)

path_trainA = path_to_datasets_scriptDir + '/trainA'
path_trainB = path_to_datasets_scriptDir + '/trainB'
path_testA = path_to_datasets_scriptDir + '/testA'
path_testB = path_to_datasets_scriptDir + '/testB'

shutil.rmtree(path_trainA) if 'trainA' in os.listdir(path_to_datasets_scriptDir) else None
shutil.rmtree(path_trainB) if 'trainB' in os.listdir(path_to_datasets_scriptDir) else None
shutil.rmtree(path_testA) if 'testA' in os.listdir(path_to_datasets_scriptDir) else None
shutil.rmtree(path_testB) if 'testB' in os.listdir(path_to_datasets_scriptDir) else None

if 'trainA' not in os.listdir(path_to_datasets_scriptDir):
    os.mkdir('trainA')
if 'trainB' not in os.listdir(path_to_datasets_scriptDir):
    os.mkdir('trainB')
if 'testA' not in os.listdir(path_to_datasets_scriptDir):
    os.mkdir('testA')
if 'testB' not in os.listdir(path_to_datasets_scriptDir):
    os.mkdir('testB')
os.chdir(path_to_datasets)


def copy2temp(src, dst):
    data_already_exist = os.listdir(dst + '/__temp')
    for file in os.listdir(src):
        if len(data_already_exist) == 0 or file not in data_already_exist:
            shutil.copy(src + f'/{file}', dst + '/__temp')

            
def clear_double_temp():
    data_already_exist = os.listdir(path_to_datasets + '/__temp')
    for file in data_already_exist:
        os.remove(f'{path_to_datasets}/__temp/{file}')

        
from PIL import Image
import sys
def resize_pics(size_num, path_in, path_out):
    for item in os.listdir(path_in):
        if os.path.isfile(path_in+f'/{item}'):
            im = Image.open(path_in+f'/{item}')
            f, e = os.path.splitext(path_out+f'/{item}')
            imResize = im.resize((size_num, size_num), Image.ANTIALIAS)
            imResize.save(f + '.jpg', 'JPEG', quality=90)

            
def move2dir(src, dst):
    data_already_exist = os.listdir(dst)
    for file in os.listdir(src):
        if len(data_already_exist) == 0 or file not in data_already_exist:
            shutil.move(src + f'/{file}', dst)
    clear_double_temp()


import random as rnd 
def get_test_files(pics_to_test, **dst_arr):
    def do(path_input, path_output):
        local_files = os.listdir(path_input)
        num_local_files = len(local_files)
        print(num_local_files)
        for i in range(pics_to_test):
            pic_for_test = rnd.choice(local_files)
            if pic_for_test not in os.listdir(path_output):
                shutil.move(path_input + f'/{pic_for_test}', path_output)
            local_files.remove(pic_for_test)
    dirname_anime = dst_arr['trainB']
    dirname_people = dst_arr['trainA']
    dst_dir_anime = dst_arr['testB']
    dst_dir_people = dst_arr['testA']    
    do(path_input=dirname_anime, path_output=dst_dir_anime)
    do(path_input=dirname_people, path_output=dst_dir_people)
            


# trainA (anime)
copy2temp(src=from_what, dst=path_to_datasets)
resize_pics(size_num=128, path_in=path_to_datasets + '/__temp', path_out=path_to_datasets + '/temp')
move2dir(src=path_to_datasets + '/temp', dst=path_trainA)

# trainB (simplified, normal)
copy2temp(src=to_what, dst=path_to_datasets)
resize_pics(size_num=size_pictures, path_in=path_to_datasets + '/__temp', path_out=path_to_datasets + '/temp')
move2dir(src=path_to_datasets + '/temp', dst=path_trainB)

# testA + testB
get_test_files(pics_to_test=200, trainA=path_trainA, trainB=path_trainB, testA=path_testA, testB=path_testB)
```

```python
def create_start_script(**kwargs):
    script_name = 'main.py'
    memory_type = '--light ' + kwargs['light'] if kwargs['light'] else " "
    iterations = '--iteration ' + kwargs['iteration'] if kwargs['iteration'] else str(200000)
    batch_size = '--batch_size ' + kwargs['batch_size'] if kwargs['batch_size'] else str(1)
    print_freq = '--print_freq ' + kwargs['print_freq'] if kwargs['print_freq'] else str(5000)
    save_freq = '--save_freq ' + kwargs['save_freq'] if kwargs['save_freq']else str(5000)
    img_size = '--img_size ' + kwargs['img_size'] if kwargs['img_size'] else str(128)
    dataset = '--dataset ' + kwargs['dataset']
    resume = '--resume ' + kwargs['resume'] if kwargs['resume'] else " "
    data_out = f'!python {script_name} {memory_type} {iterations} {batch_size} {print_freq} {save_freq} {img_size} {dataset} {resume}'
    return data_out
```

```python
create_start_script(light='true',
                    iteration=str(250000),
                    batch_size=str(5),
                    print_freq=str(5000),
                    save_freq=str(5000),
                    img_size=str(128),
                    dataset="SIMPlified",
                    resume = '')
```
